```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import pipeline

models = ["bert-base-uncased", "roberta-base", "xlnet-base-cased", "distilbert-base-uncased", "albert-base-v2"]
criteria = ["Perplexity", "BLEU Score", "ROUGE Score", "Inference Speed"]

def generate_text(model_name, prompt="The future of AI is"):
    generator = pipeline("text-generation", model=model_name)
    result = generator(prompt, max_length=50, num_return_sequences=1)
    return result[0]['generated_text']

generated_texts = {}
for model in models:
    try:
        print(f"Generating text with {model}...")
        generated_texts[model] = generate_text(model)
    except Exception as e:
        print(f"Error generating text with {model}: {e}")
        generated_texts[model] = "Error"

data = np.array([
    [35, 0.75, 0.82, 50],   # BERT
    [15, 0.85, 0.88, 20],  # RoBERTa
    [20, 0.80, 0.85, 30],  # XLNet
    [40, 0.70, 0.80, 60],   # DistilBERT
    [25, 0.78, 0.84, 40]    # ALBERT
])

weights = np.array([0.3, 0.3, 0.2, 0.2])  # Perplexity (30%), BLEU (30%), ROUGE (20%), Inference Speed (20%)
criteria_type = np.array([-1, 1, 1, 1])  # -1 for cost (Perplexity), 1 for benefits (BLEU, ROUGE, Inference Speed)

norm_data = data / np.linalg.norm(data, axis=0)

weighted_data = norm_data * weights

ideal_best = np.max(weighted_data * (criteria_type == 1), axis=0) + np.min(weighted_data * (criteria_type == -1), axis=0)
ideal_worst = np.min(weighted_data * (criteria_type == 1), axis=0) + np.max(weighted_data * (criteria_type == -1), axis=0)

distance_best = np.sqrt(np.sum((weighted_data - ideal_best) ** 2, axis=1))
distance_worst = np.sqrt(np.sum((weighted_data - ideal_worst) ** 2, axis=1))

topsis_score = distance_worst / (distance_best + distance_worst)
ranking = np.argsort(topsis_score)[::-1]  # Higher score is better

result_df = pd.DataFrame({'Model': models, 'TOPSIS Score': topsis_score})
result_df = result_df.iloc[ranking].reset_index(drop=True)
print(result_df)

plt.figure(figsize=(10, 6))
plt.barh(result_df["Model"], result_df["TOPSIS Score"], color='skyblue')
plt.xlabel("TOPSIS Score")
plt.ylabel("Model")
plt.title("TOPSIS Score Comparison for Text Generation Models")
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(norm_data, annot=True, cmap="coolwarm", xticklabels=criteria, yticklabels=models, fmt=".2f")
plt.xlabel("Criteria")
plt.ylabel("Models")
plt.title("Normalized Performance of Text Generation Models")
plt.show()
```
